## Overview

In this lab, you will use a differential privacy library to generate a synthetic version of a dataset.  You will compare the performance of a classifier trained on the original data and on the synthetic data and comment on the "cost" of privacy. 

Differential privacy allows us to add random noise to a statistical result to "hide" the individuals in the original data.   With the basic idea, we can redesign more complicated models that are also differentially private.   In this lab, we are actually using a neural network, with some carefully added random noise, to model the original data.  Once we have trained this (noisy) model, we can ask it to generate random records that look like the original records.  We might use this random, private data as training data for some other task, hopefully achieving similar results while preserving privacy.

```python
import pandas as pd
import numpy as np
from snsynth.pytorch.nn import PATECTGAN
from snsynth.pytorch import PytorchDPSynthesizer
from snsynth import MWEMSynthesizer
```

## Step 1: Load the data

The data is available on the web.   Use read_csv as usual.  Drop the income column, as PATECTGAN can only handle categorical variables.

We'll also bin the income variable so we can think about this problem as a classification problem.  We'll try to predict whether someone is under the [US poverty level for a family of four](https://aspe.hhs.gov/topics/poverty-economic-mobility/poverty-guidelines/prior-hhs-poverty-guidelines-federal-register-references/2021-poverty-guidelines).

```python
pums_csv_path = "https://raw.githubusercontent.com/opendp/dp-test-datasets/master/data/PUMS_california_demographics_1000/data.csv"
pums = pd.read_csv(pums_csv_path, index_col=None) # in datasets/

poverty = pums['income'] <= 26500  # US poverty level for a family of four

pums = pums.drop(['income'], axis=1)
pums["poverty"] = poverty
```

```python
pums
```

## Step 2: Use PATECTGAN to generate synthetic data

We indicate which method we will use, and configure some options.  We then fit the model using the PUMS dataset.

Once the model is fit, we can sample it to generate a dataset of whatever size we want.

You can try another method, or work with the parameters

```python
#synth = MWEMSynthesizer(epsilon=1.0, split_factor=pums.shape[1]) 
#synth.fit(pums)

synth = PytorchDPSynthesizer(1.0, PATECTGAN(regularization='dragan'), None)
synth.fit(pums, categorical_columns=pums.columns.values.tolist() )

sample = synth.sample(10) # synthesize 10 rows
print(sample)
```

## Step 3: Visualize the results

Compare the original data with the synthesized data.  Do they look similar enough to be useful?  (My answer: no!)

```python
import seaborn as sns
from matplotlib import pyplot as plt

g = sns.FacetGrid(pums, row="sex", col="race", margin_titles=True)
g.map(sns.scatterplot, "age", "educ", "poverty", alpha=.7)
plt.figure()

sample = synth.sample(1000) # synthesize 1000 rows

g = sns.FacetGrid(sample, row="sex", col="race", margin_titles=True)
g.map(sns.scatterplot, "age", "educ", "poverty", alpha=.7)
```

## Step 4: Train a model on each and compare results

We will split the original data into train and test.

Then we will train two classifiers: one on the original training data, and one on some synthesized trainind data. 

Finally, we evaluate on the held out test data.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import ensemble

X = pums.drop(['poverty'], axis=1)
y = poverty

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=21)

```

Train and evaluate on original data

```python
original = ensemble.GradientBoostingClassifier()
original.fit(X_train, y_train)

original.score(X_test, y_test)
```

Train on synthetic data, but still evaluate on original data

```python
sample = synth.sample(1000) # synthesize 1000 rows

synth_y = sample['poverty']
synth_X = sample.drop(['poverty'], axis=1)


synthetic = ensemble.GradientBoostingClassifier()
synthetic.fit(synth_X, synth_y)

synthetic.score(X_test, y_test)
```

Not so good!!! Training on sample data but evaluating on original test set is much much worse.
